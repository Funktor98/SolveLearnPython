```python
s = 'Simple is better than complex'
print(s)
print()

# printing second char and the last one.
print('Second character is \t\'' + s[2] + '\'')
print('last one is \t\t\'' + s[-1] + '\'')

print('s[0:10]: \t\t\'' + s[0:10] + '\'')
print('s[:8]: \t\t\t\'' + s[:8] + '\'')
print('s[-7:-2]: \t\t\'' + s[-7:-2] + '\'')
print('s[-4:]: \t\t\'' + s[-4:] + '\'')  # feel the difference with s[-4] !
print('s[9:-1]: \t\t\'' + s[10:-1] + '\'')

print('s[::2] = \t\t ' + s[::2])
print('s[1::2] = \t\t ' + s[1::2])
print('s[::-1] = \t\t ' + s[::-1])
print('s[5::-1] = \t\t ' + s[5::-1])
```

##    . find, rfind, count.
1.       ( )        **in**. <br />
 *'ab' in 'abcdefdef' = True*
1.    ,        **ch**,    **find**: *my_string.find(ch)*. <br />
         (. )
2.     ,  ,  **rfind**.
4.     -     **count**.         , . .

```python
s = 'abcdefdefaaa'
print("'ab' in 'abcdefdefaaa' is " + str('ab' in s))
print("'ac' in 'abcdefdefaaa' is " + str('ac' in s))
print()

print('searching \'d\' from the left side: ' + str(s.find('d')))
print('searching \'d\' from the right side: ' + str(s.rfind('d')))
print()

print('counting a in ' + s + ': ' + str(s.count('a')))
print('counting aa in ' + s + ': ' + str(s.count('aa')))
print('(last one worked somewhat wrong)')
```

##     . replace, format.

1.     ch_old   ch_new,    **replace(old_char, new_char)**
2.           **str()** (     ,       )   **format**.       ,  .  -    {0}, {1}, ...    : **pattern_string.format(param1, param2, ...)**.      (. ).

```python
s = 'potatos,eggs,milk,salt,tomatos'
t = '{0} is the first parameter in this pattern, while {1} is the second one.'
name = 'John'
age = 57

print(s)
print()
print(s.replace(',', ' '))
print(s.replace('to', 'TO'))
print()

print(t.format('param1', 'param2'))
print('My name is {0}. I am {1} years old.'.format(name, age))
# instead of print('My name is ' + name + 'I am ' + str(age) + 'years old.')
```
